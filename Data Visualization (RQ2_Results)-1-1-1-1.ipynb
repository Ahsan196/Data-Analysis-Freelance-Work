```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

# Decision Tree

```python
df = pd.read_excel ('RQ2_Results.xlsx',sheet_name='Sheet4',index_col=0,header=1)
print (df)
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header
df.head(5)
```

```python
print(df.columns)
```

```python
x=df.columns
a = df.loc['Firefox']
b = df.loc['Core']
c=df.loc['Mac']
d=df.loc['Windows']
e=df.loc['Chrome']
f=df.loc['Android']
g=df.loc['NetBeans']

plt.ylim(0, 1.1)


plt.plot(x, a, 'b-o', label='Firefox')
plt.plot(x, b, 'g-o', label='Core')
plt.plot(x, c, 'r-o', label='Mac')
plt.plot(x, d, 'c-o', label='Windows')
plt.plot(x, e, 'm-o', label='Chrome')
plt.plot(x, f, 'y-o', label='Android')
plt.plot(x, g, 'k-o', label='NetBeans')

plt.title("Decision Tree")
plt.legend(loc='upper left')
plt.xlabel('')
plt.ylabel('F1-Score')
```

```python
dataframes = [df,df1,df2,df3,df4]
y_values = ["Firefox","Core","Mac","Windows","Chrome","Android","NetBeans"]
methods = ["Decision Tree","Random Forest","Decision Jungle","Neural Network","SVM"]
fig = plt.figure(figsize=(18,10))
plt.subplots_adjust(wspace = 0.3)

for i in enumerate(dataframes):
    x = i[1].columns
    number = i[0]
    #print(x)
    ax = fig.add_subplot(2,3,i[0]+1)
        #print(y)
    plt.plot(x,i[1].loc[y_values[0]],label=y_values[0],marker="o")
    plt.plot(x,i[1].loc[y_values[1]],label=y_values[1],marker="o")
    plt.plot(x,i[1].loc[y_values[2]],label=y_values[2],marker="o")
    plt.plot(x,i[1].loc[y_values[3]],label=y_values[3],marker="o")
    plt.plot(x,i[1].loc[y_values[4]],label=y_values[4],marker="o")
    plt.plot(x,i[1].loc[y_values[5]],label=y_values[5],marker="o")
    plt.plot(x,i[1].loc[y_values[6]],label=y_values[6],marker="o")

    plt.yticks(np.arange(0,1.1,0.1))
    plt.xlabel('')
    plt.title(methods[number])
    plt.ylabel('F1-Score')
    if i[0] == 4:
        plt.legend(bbox_to_anchor=(1.7, 1), loc='upper right', ncol=1)
    
    
```
