```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

# no-seybel-send Data

# 1. Vizualisation of the number of parcels delivered throughout the year


```python
dtf = pd.read_csv(r'E:\to send revision\Rev Notebook CSV Files\Number of Parcels Per Day.csv',header=0)
dtf.head(5)
```

```python
fig = plt.figure(figsize = (200, 8))
ax = fig.add_subplot()
labels = ax.get_xticks().tolist()
labels = pd.to_datetime(labels)
ax.set_xticklabels(labels, rotation=90)
ax=sns.barplot(x=dtf['Delivery Date'],y=dtf['Parcels Per Day'],data=dtf,palette='rainbow')
ax.set_title("Number of parcels delivered each day throughout the year", fontsize = 15)
```
